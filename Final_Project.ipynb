```python
#Count by INSPECTION_TYPE

#Initial inspection is the highest type of inspections overall, next to compliance inspections


df2['INSPECTION_TYPE'].value_counts()

#visualization of data by INSPECTION_TYPE

df2['INSPECTION_TYPE'].value_counts() \
.head() \
.plot(kind='bar', xlabel = 'INSPECTION_TYPE', rot=45,
      ylabel = 'Overall count', color = 'coral')
plt.title('INSPECTION_TYPE Comparison on Overall count' \
          , loc='center', color='olive', style='italic')
plt.show()
```

```python
# distribution, by INSPECTION_TYPE + RESULT + BOROUGH (IRB)

grouping_by_IRB = df2.groupby(['INSPECTION_TYPE', 'RESULT', 'BOROUGH']).size().reset_index(name='count')

grouping_by_IRB
```

```python
#########################
#########################
#####Manhattan has the highest count in baits applied, while Brooklyn is 2nd runner up.
#####Brooklyn has highest count in compliance inspection, followed by Bronx
#Brooklyn has the highest number of farmers market locations, next is Manhattan
#Brooklyn has the highest count of inspection monitoring visits
#Manhattan has the highest count of Markets open year round, next to Brooklyn
#Bronx has highest count of cooking demonstration, next to Brooklyn
# Bronx has the highest count in failed results and rat activities, Brooklyn is 2nd highest
# Bronx has the highest count in failed results on Initial inspection
#Bronx has the highest count in failed results and rat activities, under compliance inspection
#####Manhatan has the highest count in initial inspection,
#####Staten Island has the least count overall

# distribution, by INSPECTION_TYPE + BOROUGH (IB)
#
#

#########################
#####
#####



grouping_by_IB = df2.groupby(['INSPECTION_TYPE', 'BOROUGH']).size().reset_index(name='count')

grouping_by_IB

#visualization of data by INSPECTION_TYPE + BOROUGH (IB)

plt.figure(figsize=(7,5))
sns.barplot(x='INSPECTION_TYPE',y='count',palette='Accent', hue='BOROUGH' \
            , data=grouping_by_IB,color='coral_r')
plt.title('$\mathbf{visualization of data by IB}$' \
          , loc='center', color='coral')
plt.xlabel('INSPECTION_TYPE')
plt.ylabel('Total Count')
plt.legend(loc='best')

plt.show()
```

# Exploring Farmers Market Locations data set

```python
#Manhattan has the hight county of inspection by Borough and next is Bronx
#Brooklyn has the highest number of farmers market locations, next is Manhattan then Bronx
#Brooklyn is the 3rd highest and Staten Island has the least count

#####Manhattan has the highest count in baits applied, while Brooklyn is 2nd runner up.
#####Brooklyn has highest count in compliance inspection, followed by Bronx
#Brooklyn has the highest number of farmers market locations, next is Manhattan
#Brooklyn has the highest count of inspection monitoring visits
#Manhattan has the highest count of Markets open year round, next to Brooklyn
#Bronx has highest count of cooking demonstration, next to Brooklyn
# Bronx has the highest count in failed results and rat activities, Brooklyn is 2nd highest
# Bronx has the highest count in failed results on Initial inspection
#Bronx has the highest count in failed results and rat activities, under compliance inspection
#####Manhatan has the highest count in initial inspection,
#####Staten Island has the least count overall



df_B_1['Borough'].value_counts() \
.head(10) \
.plot(kind='bar', xlabel = 'Borough', rot=45,
      ylabel = 'Overall count', color = 'coral')
plt.title('Borough Comparison on Overall count' \
          , loc='center', color='olive', style='italic')
plt.show()

```

```python
###Visualization of Borough + Open Year-Round (BOYR)
# Manhattan has the highest count of Farmers markets open year round and Brooklyn is 2nd highest
#Manhattan has the highest count of Rodent inspections overall + highest count of Rat baits applied
# Manhattan has the highest count of Rat activities on initial inspection and Bronx coming up 2nd
#Bronx has the highest count in failed results and rat activities, under compliance inspection
#Brooklyn has the highest count in monitoring inspection visits.
#The county with the least count of Farmers market opened year round is Staten Island

grouping_by_BOYR = df_B_1.groupby(['Borough', 'Open Year-Round']).size().reset_index(name='count')

grouping_by_BOYR

plt.figure(figsize=(7,5))
sns.barplot(x='Borough',y='count',palette='Blues', hue='Open Year-Round' \
            , data=grouping_by_BOYR,color='BrBG_r')
plt.title('$\mathbf{Borough Comparison on Open Year-Round}$' \
          , loc='center', color='coral')
plt.xlabel('Borough')
plt.ylabel('Total Count')
plt.legend(loc='upper right')

plt.show()
```

```python
# distribution, by Borough +  Cooking Demonstrations (BCD)

grouping_by_BCD = df_B_1.groupby(['Borough', ' Cooking Demonstrations']).size().reset_index(name='count')

grouping_by_BCD

###Visualization of Borough + Cooking Demonstrations (BCD)
# Bronx has the highest count of locations with cooking demonstration, next was Brooklyn
# Bronx has the highest count in failed results and rat activities
# Bronx has the highest count in failed results on Initial inspection
#Bronx has the highest count in failed results and rat activities, under compliance inspection
#Staten Island had the least count of cooking demonstration locations

plt.figure(figsize=(7,5))
sns.barplot(x='Borough',y='count',palette='pink_r', hue=' Cooking Demonstrations' \
            , data=grouping_by_BCD,color='Blues')
plt.title('$\mathbf{Borough Comparison on Cooking Demonstrations}$' \
          , loc='center', color='coral')
plt.xlabel('Borough')
plt.ylabel('Total Count')
plt.legend(loc='upper right')

plt.show()
```

# Merging both data sets

```python
#joing my 2 different dataframes to make a new one

df_new = pd.concat([df2, df_B_1], keys=["Rod", "Mar"])

df_new.head()
```

```python
#Number of rows and columns

df_new.shape
```

```python
#checking to see if there are missing values in my dataframe

df_new.isnull().sum()
```

```python
#Replacing Nan values---- Cleaning Data

df_new.fillna(0)

df_new[['ZIP_CODE', 'Zip Code', 'JOB_PROGRESS']] = df_new[['ZIP_CODE', 'Zip Code', 'JOB_PROGRESS']].fillna(0)

df_new[['Borough', 'BOROUGH', 'RESULT', 'Market Name']] = df_new[['Borough', 'BOROUGH', 'RESULT', 'Market Name']].fillna('None')

df_new[['Open Year-Round', ' Cooking Demonstrations', 'INSPECTION_TYPE']] = df_new[['Open Year-Round', ' Cooking Demonstrations', 'INSPECTION_TYPE']].fillna('No')

df_new
```
